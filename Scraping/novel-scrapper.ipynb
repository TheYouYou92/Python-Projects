```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time
import csv
import asyncio

def to_csv(res):
    with open("novel.csv", "a",newline='', encoding='utf-8') as f:
        writer = csv.DictWriter(f, fieldnames=["novelname", "title","chapternum","image","discription","tag","author","category","text"], delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
        writer.writerow(res)

options = webdriver.ChromeOptions()
        # Add the headless argument
options.add_argument("--headless")
        # Add other arguments to improve performance
options.add_argument("--disable-gpu")
options.add_argument("--no-sandbox")
options.add_argument("--window-size=1920,1080")
driver = webdriver.Chrome(options=options)
driver.get('https://www.69shu.pro/all.html')

html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
time.sleep(1)
urls = soup.find_all('a')
async def novels():
        for x in urls:
                if '/book/' in x['href']:
                        dic = {}
                        dic['novelname'] = x.text.strip()
                        url = 'https://www.69shu.pro' + x['href']
                        driver.get(url)
                        nohtml = driver.page_source
                        nosoup = BeautifulSoup(nohtml, 'html.parser')
                        dic['image'] = nosoup.find('div', {'class':'bookimg2'}).img['src']
                        dic['discription'] = nosoup.find('div',{'class':'navtxt'}).get_text().strip()
                        try:
                                dic['tag'] = nosoup.find('ul',{'class':'tagul'}).get_text(separator=',')
                        except:
                                dic['tag'] = ' '
                        booknav = nosoup.find('div',{'class':'booknav2'})
                        tags = booknav.findAll('p')
                        dic['author'] = tags[0].a.text
                        dic['category'] = tags[1].a.text
                        driver.get(url.replace('.htm','/'))
                        chahtml = driver.page_source
                        chasoup = BeautifulSoup(chahtml, 'html.parser')
                        chaurls = chasoup.find_all('li')
                        for x in chaurls:
                                if '/txt/' in x.a['href']:
                                        dic['title'] = x.a.text.strip()
                                        url = x.a['href']
                                        dic['chapternum'] = x['data-num']
                                        driver.get(url)
                                        time.sleep(1)
                                        chtml = driver.page_source
                                        csoup = BeautifulSoup(chtml, 'html.parser')
                                        dic['text'] = csoup.find('div',{'class':'txtnav'}).getText().strip()
                                        to_csv(dic)
                                        break
asyncio.run(novels())
```
