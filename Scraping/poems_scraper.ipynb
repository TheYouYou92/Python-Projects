```python
from selenium import webdriver
import threading
import time
from bs4 import BeautifulSoup
import pandas as pd

# Define the base url and the poets page
base_url = 'https://mypoeticside.com'
poets_page = base_url + '/poets'

# Define the remote server url and the browser options

options = Options()
options.add_argument("--headless")
driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()),options=options)
# Define the number of browser instances and threads to create
num_browsers = 10

# Create an empty list to store the browser instances and the threads
browsers = []
threads = []

# Define a function that takes a browser instance and a page number as parameters
def scrape_page(browser, page):

    # Get the current page url
    page_url = poets_page + '/page/' + str(page)

    # Go to the current page
    browser.get(page_url)

    # Wait for the page to load
    time.sleep(3)

    # Get the html source of the page
    page_source = browser.page_source

    # Parse the html using BeautifulSoup
    soup = BeautifulSoup(page_source, 'html.parser')

    # Find all the poets in the page
    poets = soup.find_all('h3')

    # Loop through each poet in the page
    for poet in poets:

        # Get the poet's name
        poet_name = poet.a.text.strip()

        # Get the poet's url
        poet_url = poet.find('a')['href']

        # Go to the poet's page
        browser.get(poet_url)

        # Wait for the page to load
        time.sleep(3)

        # Get the html source of the page
        poet_source = browser.page_source

        # Parse the html using BeautifulSoup
        poet_soup = BeautifulSoup(poet_source, 'html.parser')

        # Find all the poems of the poet
        poems = poet_soup.find_all('a', href=lambda x: x and 'show-classic-poem' in x)
        poems_links = []
        # Create an empty string to store the poems
        poet_poems = ''
        

        # Loop through each poem of the poet
        for poem in poems:
            
            # Get the poem's title
            poem_title = poem.text.strip()

            # Get the poem's url
            poem_url = 'https://' + poem['href'][2:]

            # Go to the poem's page
            browser.get(poem_url)

            # Wait for the page to load
            time.sleep(3)

            # Get the html source of the page
            poem_source = browser.page_source

            # Parse the html using BeautifulSoup
            poem_soup = BeautifulSoup(poem_source, 'html.parser')
            try:
                # Find the poem's text
                poem_text = poem_soup.find('p').text.strip()

                # Add the poem's title and text to the poet's poems
                poet_poems += "<h2>"+ poem_title +"</h2>"  + '\n' +  poem_text + '\n\n'
            except:
                continue
        # Add the poet's name and poems to the list
        poets_and_poems.append((poet_name, poet_poems))

# Define the total number of pages of poets
total_pages = 280

# Define a counter for the current page
current_page = 1

# Loop until all pages are scraped
while current_page <= total_pages:

    # Loop through the number of browser instances
    for i in range(num_browsers):

        # Check if the current page is within the total pages
        if current_page <= total_pages:

            # Create a new browser instance and connect to the remote server
            browser = webdriver.Remote(command_executor=driver, options=options)

            # Append the browser instance to the list
            browsers.append(browser)

            # Create a new thread and pass the browser instance and the current page as arguments
            thread = threading.Thread(target=scrape_page, args=(browser, current_page))

            # Append the thread to the list
            threads.append(thread)

            # Start the thread
            thread.start()

            # Increment the current page by one
            current_page += 1

        # Otherwise, break the loop
        else:
            break

    # Wait for all threads to finish
    for thread in threads:
        thread.join()

    # Close all browser instances
    for browser in browsers:
        browser.close()

    # Clear the lists
    browsers.clear()
    threads.clear()

# Print the list of poets and their poems
df = pd.DataFrame(poets_and_poems)
df.to_csv('all_poems.csv')
```
